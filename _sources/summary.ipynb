## Have We Made Any Progress?
Given all the above, it might be natural to wonder whether we have actually made any progress over the repeated measures ANOVA? Although we have an estimation framework that is more general, and we have removed the awkwardness of having to manually construct inferential tests via multiple ANOVA tables, the lack of sphericity correction means that we are tied-in to compound symmetry. 

- Sphericity is misleading in practice?
- The data do not support more complex estimation, but it could if we had replications

However, the true of the matter is that there are some situations where a GLS model may be superior, for our purposes...

## What to Use?
... If you have a simple repeated measurement dataset with *no repeats* (i.e. each subject only has *one measurement* per repeated measure), then your options are (in order of *least preferential* to *most preferential*):

1. The traditional repeated measures ANOVA
    - Compound symmetric covariance structure and difficult to fit correctly using `aov()` with an `Error()` term. 
    - Advantage of *inferential certainty*, as all the theory from the normal linear model still applies.
    - Even when the theory does *not* apply, effective degrees of freedom can be calculated via sphericity correction.
    - Use `ezANOVA()` to fit, but recognise that you are throwing-away the linear models frameowork by doing so.
2. A mixed-effects model
    - Using a *random intercept* model, the covariance structure remains compound symmetric, which is no better than the repeated measures ANOVA.
    - Additional *inferential uncertainty*, as the theory from the normal linear model no longer applies.
    - No sphericity correction, though *effective degrees of freedom* are available alongside *asymptotic* results.
    - No need to specify error terms, as these are accommodated automatically.
    - Remains within the linear models framework, so we can still get residuals and diagnostic plots as usual.
3. Generalised Least Squares
    - Freedom to estimate a completely unconstrained covariance structure.
    - Greater *inferential uncertainty* as the additional structure from the data is not an inherent part of the model and cannot be used as easily to infer effective degrees of freedom. This largely forces us into *asymptotic* results.
    - Error terms remain automatically accommodated.
    - Like mixed-effects, remains within the linear models framework.

So, in this situation, GLS may be the better approach. This is quite an important conclusion because many practising researchers in psychology have moved to mixed-effects models as their *default* replacement for the repeated measures ANOVA. However, under certain circumstances, GLS can lift more assumptions than a mixed-effects model. 

## When are Mixed-effects Models Useful?
... The problem is that we have not really seen the utility of mixed-effects models in the current examples. Next week, we will introduce a variety of data structures that mixed-effects models can be applied to. In doing do, we will see how flexible the framework is and why it is so useful. 

For now, there is an important lesson in all of this: the utility of a mixed-effects models does *not* come from data with no replications. In fact, there is little that a mixed-effects model can do here because the data *does not support* a more complex random-effects structure. This is key not only from an *analysis* perspective, but also from a *design* perspective. The problem is that many researchers are taught to used mixed-effects models as a drop-in replacement for the repeated measures ANOVA, yet this application effectively *neuters* the mixed-effects framework. Arguably, GLS should be used instead because it can relax the covariance assumptions entirely, even with no replications per-repeat. 

The catch is that almost all experiments in experimental and clinical psychology *do* have some form of replication within the repeated measurements. All neuropsychological tasks will have multiple trials per-condition. For instance, even a basic Stroop task will have multiple *congruent* and *incongruent* trials. Psychologists are often in the habit of *throwing this information away* by either summarising the trials through total scores, or creating condition averages per-subject. This is partly because this information cannot be used in traditional repeated measures analyses. However, as we will see in the materials next week, such information is *essential* for mixed-effects models to operate. So, if you do not have it, a mixed-effects model may not be the best choice. However, if you *do* have it or are able to *recover* it, a mixed-effects model suddenly becomes viable again.
