# Summary
In this lesson, we have introduced the mixed-effects framework as an alternative to both traditional repeated measurement models and as an alternative to GLS. We have seen how to represent a model in a multilevel fashion, building up complexity in layers, before collapsing down to mixed-effects form. Furthermore, we have seen how the mixed-effects framework does not require pre-specification of a covariance structure. Instead, the structure becomes a natural emergent property of the model. This alone is an advantage. However, it is difficult to overlook the fact that a mixed-effects model is a fundamentally more complex affair than either the repeated measures ANOVA or a GLS model. Indeed, if our aim is to get away from assumptions of compound symmetry and the need for manual specification of error terms, the examples in this lesson are all suggestive of using GLS rather than a mixed-effects model. So, from our current vantage, it might be natural to wonder whether we have actually made any progress at all?

The problem is that we have not really seen the utility of mixed-effects models in the current examples. Simple repeated measurement datasets with no replications of the repeats are not where mixed-effects models really shine. Next week, we will introduce a variety of data structures that mixed-effects models can be applied to. In doing do, we will see how flexible the framework is and why it is so useful. For now, there is an important lesson in all of this: the utility of a mixed-effects models does *not* come from data with no replications. In fact, there is little that a mixed-effects model can do here because the data *does not support* a more complex random-effects structure. This is key not only from an *analysis* perspective, but also from a *design* perspective. The problem is that many researchers are taught to used mixed-effects models as a drop-in replacement for the repeated measures ANOVA, yet this application effectively *neuters* the mixed-effects framework. Arguably, GLS should be used instead because it can relax the covariance assumptions entirely, even with no replications per-repeat. 

The catch is that almost all experiments in experimental and clinical psychology *do* have some form of replication within the repeated measurements. Indeed, almost all experiments will have *multiple trials* of the same condition. The problem is that psychologists are often in the habit of *throwing this information away* by either summarising the trials through total scores, or creating condition averages per-subject. This is why the datasets in the `datarium` package are the way that they are. It is almost a certainty that the experiments that generated many of these datasets had multiple trials, but this information has been discarded. As we will see in the materials next week, such information is *essential* for mixed-effects models to operate. So, if you do not have it, a mixed-effects model may not be the best choice. If you *do* have it, a mixed-effects model suddenly becomes viable again. More so, a mixed-effects models can become the *better choice*.
