## The Hierarchical Framework

$$
\begin{alignat*}{1}
    y_{is} &= \mu_{s} + \epsilon_{is} \\
    \mu_{s} &= \mu + \eta_{s}
\end{alignat*}
$$

$$
\begin{alignat*}{1}
    y_{is} &= \mu + \eta_{s} + \epsilon_{is} \\
    \eta_{s} &\sim \mathcal{N}\left(0,\sigma^{2}_{b}\right) \\
    \epsilon_{is} &\sim \mathcal{N}\left(0,\sigma^{2}_{w}\right) \\
\end{alignat*}
$$
