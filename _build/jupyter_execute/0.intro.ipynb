# Introduction
In this lesson, we will introduced mixed-effects models. As we have now seen, traditional methods of analysing repeated measurements are limited by the normal linear model framework. Last week, we discussed a possible solution in the form of Generalised Least Squares (GLS). Although GLS has many useful applications, it is fundamentally limited by the need to specify the covariance structure up-front. This can prove challenging, especially with complex data. Furthermore, GLS has no sense of the *structure* of the data, particularly in terms of *where* correlation comes from. This means it is unable to use that structure to its advantage. Mixed-effects models solve both of these issues. By focussing on the structure of the data, we build models that represent where our data come from. Rather than deciding on a covariance structure up-front, the covariance structure becomes an emergent property of the model. As such, for data with many complex dependencies, our only challenge is specifying the data structure and a suitable variance-covariance matrix will simply emerge. Furthermore, by embedding the data structure in the model, mixed-effects models are able to accommodate features such as missing data much more elegantly. For all these reasons, mixed-effects models have become a staple of modern statistics. You will seem them used widely used across many different disciplines. However, these methods are inherently *complex* and often not fully understood. Across the next two weeks, we will aim to introduce the use and application of these approaches slowly and carefully, with the broad aim of making them as intuitive as possible. That being said, it is worth understanding from the start that these methods fundamentally require more thought and consideration than simply picking a covariance structure and getting on with it.
